#**Playing Activity Will Be Carried Out Or Not. Based Weather Prediction Using Naive Bayes, Logistic Regression, and KNN Classification**

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import GaussianNB
```

```python
# reading CSV file
df = pd.read_csv('weather.csv')
df
```

```python
#Encoding strings to numeric
num = LabelEncoder()

#Dropping the target variable and make it is as newframe
inputs = df.drop('play', axis = 'columns')
target = df['play']
target
```

```python
#Creating a new dataframe
inputs['outlook_n'] = num.fit_transform(inputs['outlook'])
inputs['temp_n'] = num.fit_transform(inputs['temperature'])
inputs['humidity_n'] = num.fit_transform(inputs['humidity'])
inputs['windy_n'] = num.fit_transform(inputs['windy'])
inputs
```

```python
inputs_n = inputs.drop(['outlook','temperature','humidity','windy'],axis = 'columns')
inputs_n
```

## **Gaussian NaiveBayes Clasification**

```python
#Applying Gaussian Naivebayes
classifier = GaussianNB()
classifier.fit(inputs_n,target)
```

```python
#Predicting Accuracy
classifier.score(inputs_n,target)
```

```python
#Prediction
classifier.predict([[2,1,1,1]])
```

##**Logistic Regression and KNN Classification**

```python
X = inputs_n
Y = target
```

##**Train, Test and Split**

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y)
```

```python
print(x_train.shape,"......",x_test.shape)
print(y_train.shape,"......",y_test.shape)
```

###**Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
```

```python
LogisticClassifier = LogisticRegression()
```

```python
#Training Model on Training Data
LogisticClassifier.fit(x_train,y_train)
```

###*Evaluation of Accuracy*

```python
from sklearn.metrics import accuracy_score
```

```python
LogisticClassifier.score(x_train,y_train)
```

```python
LogisticClassifier.score(x_test,y_test)
```

```python
#Prediction
LogisticClassifier.predict([[2,1,1,1]])
```

##**KNN Classification**

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors = 5)
```

```python
knn.fit(x_train, y_train)
```

###Evaluation of Accuracy

```python
knn.score(x_train,y_train)
```

```python
knn.score(x_test,y_test)
```

```python
#Prediction
knn.predict([[2,1,1,1]])
```
